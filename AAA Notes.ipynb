name: greatest_common_divisor

Input params: a,b : natural numbers

Return: largest natural number that divides into both a and b

````python
def greatest_common_divisor(a,b):
  candidate_gcd = 1
  for j in range(1, min(a,b)+1):
    if (a % j==0) and (b % j == 0):
      candidate_gcd = j
  return candidate_gcd

greatest_common_divisor(24,36)

````

# **Example**

Specs:

Name: Prime factorization

Input params: n: a natural number greater than 1

Returns: list of prime numbers whose product equals n

````python
def prime_factorization(n):

  factors = []
  candidate_factor = 2
  remaining_factor = n

  while remaining_factor > 1:

    if (remaining_factor % candidate_factor == 0):
      remaining_factor = remaining_factor // candidate_factor
      factors.append(candidate_factor)
    else:
      candidate_factor = candidate_factor + 1

  return factors


prime_factorization(37)
````

````python
def best_conductivity_optimized(v):
  a, b, bestconduct, besta, bestb = 0,0,0,0,0

  factorsa = []
  factorsb = []
  factors = []

  for i in range(0, v+1):
    factorsa.append(i)

  for o in reversed(range(0, v+1)):
    factorsb.append(o)

  factors = list(zip(factorsa, factorsb))

  for i in range(len(factors)):
    a = factors[i][0]
    for o in range(len(factors)):
      b = factors[o][1]
      if a + b == v:
        if (10**10 - (a - 33)**2 - 2*(b - 23)**2) > bestconduct:
          bestconduct = (10**10 - (a - 33)**2 - 2*(b - 23)**2)
          besta = a
          bestb = b
  return besta, bestb

best_conductivity_optimized(179)
````

# Systems of difference equations




System of difference equations for simulating a harmonic oscilator
---------------------



Let $t_0 := 0$, $ y_0 := 0 $ $v_0 := 1$. 

* Let $t_n$ denote the time at step $n$ of the simulation.
* Let $y_n$ denote the position of the bob at step $n$ of the simulation.
* Let $v_n$ denote the velocity of the bob at step $n$ of the simulation.
* Let $\Delta t :=0.01 > 0$ denote the amount of time passing between steps of the simulation. 
* Let $s := 1$ denote the springiness of the spring.

Set up a system of difference equations to simulate a harmonic oscilator. 


For $n=1,2,3...$

$$ t_n := t_{n-1} + \Delta t $$
$$ y_n := y_{n-1} + \Delta t  v_{n-1} $$
$$ v_n := v_{n-1} + \Delta t s (-y_{n-1}) $$

Question 1: Function def
------------
Define a function according to the following specification
```
    Name:
        simulation_step
        
    Input parameters:
        state: A tuple of numbers 
                (t_(n-1), y_(n-1), v_(n-1))

    Return:
        A tuple of numbers (t_n, y_n, v_n)
        As determined by the above 
        system of difference equations
        from (t_(n-1), y_(n-1), v_(n-1))


```



````python
dt = 0.01
s = 1

def simulation_step(state):
  current_t, current_y, current_v = state

  next_t = current_t + dt
  next_y = current_y + dt * current_v
  next_v = current_v + dt * s * (-current_y)
  
  return (next_t, next_y, next_v)
````

Below we present a version of the Lotka-Volterra equations to model the population dynamics of a predator species and a prey species over time.

Let $y_n$ and $z_n$ respectively denote the number of prey and predators present in the environment after $n$ time units.

Let $y_0 := 200$ and $z_0 := 250$ and for $n=1,2,3,...$ define 
$$
    y_n := 1.3 y_{n-1} - 0.002y_{n-1}z_{n-1}
